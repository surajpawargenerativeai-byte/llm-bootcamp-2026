```python
import torch
from torch import nn


class TinyTokenModel(nn.Module):

    def __init__(self, vocab_size, n_embd, n_hidden):
        super().__init__()

        # Converts integer token IDs into vectors.
        self.token_embedding = nn.Embedding(
            num_embeddings=vocab_size,
            embedding_dim=n_embd,
        )

        # A tiny feed-forward network.
        self.mlp = nn.Sequential(
            nn.Linear(n_embd, n_hidden),
            nn.ReLU(),
            nn.Linear(n_hidden, vocab_size),
        )

    def forward(self, idx):

        # idx shape:
        # (B, T)

        x = self.token_embedding(idx)

        # x shape:
        # (B, T, C)

        logits = self.mlp(x)

        # logits shape:
        # (B, T, V)

        return logits


# --------------------------------------------------
# Tiny vocabulary
# --------------------------------------------------

stoi = {
    "I": 0,
    "am": 1,
    "Suraj": 2,
    "Pawar": 3,
    "and": 4,
    "study": 5,
    "until": 6,
    "4": 7,
}


# --------------------------------------------------
# Tokenized sentence
# --------------------------------------------------

idx = torch.tensor([
    [0, 1, 2, 3, 4, 0, 5, 6, 7, 1]
])


# --------------------------------------------------
# Create model
# --------------------------------------------------

model = TinyTokenModel(
    vocab_size=8,  # V
    n_embd=4,      # C
    n_hidden=6,    # H
)


# --------------------------------------------------
# Forward pass
# --------------------------------------------------

logits = model(idx)

print("idx shape:", idx.shape)
print("logits shape:", logits.shape)


# --------------------------------------------------
# Convert logits to probabilities
# --------------------------------------------------

probabilities = torch.softmax(logits, dim=-1)


# --------------------------------------------------
# Pick highest-probability token at each position
# --------------------------------------------------

predicted_ids = probabilities.argmax(dim=-1)

print("predicted IDs shape:", predicted_ids.shape)
```
